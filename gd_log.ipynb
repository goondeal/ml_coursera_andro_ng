```python
def get_min_x(start, f, threshold=0.0001):
    ''' 
        This function returns optimal x value that makes function [f] minimum with complexity ~= lg(n).
        
        [start] is the first estimation, can be any non-zero real value,
        [f] is the function applied to x, i.e. f(x) = x** + x + 2 
    '''
    
    x = [start, 1.5*start, 2*start]
    last_mid = start
    i = 0
    # The counter i is for monitoring iterations and preventing infinite loops as it is testing stage.
    # Later it will be replaced with the stopping conditions bellow.
    while i < 100:
        # Calculate y values that corresponding to x values.
        y = [f(k) for k in x]
        
        # Todo: remove logging print statements.
        i += 1
        print('i =', i)
        print('x =', x)
        print('y =', y)
        print()
        
        # Stopping condition: if all the 3 points nearly point to the same value.
        if abs(x[1] - last_mid) <= threshold and abs(x[0] - x[2]) <= 2*threshold:
            break
            
        last_mid = x[1]
        
        # IF the line forms the three points result a down arrow, move in the same direction (down).
        if y[0] < y[1] < y[2]:
            # The last x (x[2]) will take the value of the previous middle x (x[1]),
            # Middle x (x[1]) will take the value of the previous first x (x[0]),
            # First x (x[0]) will move as twice distance as the last distance between the previous first and middle x values.
            x = [x[0]-2*(x[1]-x[0]), x[0], x[1]]
        
        # IF the line forms the three points result an up arrow, move in the opposite direction (down).
        elif y[0] > y[1] > y[2]:
            x = [x[1], x[2], x[2]+2*(x[2]-x[1])]
        
        # IF the line forms the three points result a down-up arrow, fasten the middle point and pull the others towards it (down).
        else:
            avg1 = (x[1] + x[0]) / 2
            avg2 = (x[1] + x[2]) / 2
            x = [avg1, x[1], avg2]
            
    return x[1]

print(get_min_x(4, lambda x: x**2 - 8*x + 18)) # = 4
print(get_min_x(1000, lambda x: x**2)) # = 0.0
print(get_min_x(-20, lambda x: x**2 + 10*x + 23)) # = -5.0
```
